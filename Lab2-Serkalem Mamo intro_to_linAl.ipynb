```python
import numpy as np

A=np.array([[10,95,8],
   [95,20,9],
   [8,9,30]])
B=np.array([[5,8,3,6],
   [95,20,9,3],
   [8,9,30,3],
   [8,9,30,3]])
C=np.array([[0,6,250],
   [-6,0,86.3],
   [-250,-86.3,0]])

print("A=", "\n", A,"\n","B=", "\n", B,"\n", "C=", "\n", C,"\n")
```

```python
#Transpose of A,B and C
print("A Transpose:", "\n", np.transpose(A))
print("B Transpose:", "\n", np.transpose(B))
print("C Transpose:", "\n", np.transpose(C))
```

```python
#Inverse of A,B and C 
print("A Inverse:", "\n", np.linalg.inv(A))
#print("B Inverse:", "\n", np.linalg.inv(B)) #this is an invertable metrix so it's undeterminable
print("C Inverse:", "\n", np.linalg.inv(C))
```

```python
#B is a singular metrix because the determinant of the matrix is zero
print(np.linalg.det(A),np.linalg.det(B),np.linalg.det(C))
#print(np.linalg.inv(B)) # The Determinant for B is 0 so it's True Singular Matrix
```
